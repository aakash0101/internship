# Q3: Write a python program to access the search bar and search button on images.google.com and scrape 10 images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’, ‘Guitar’, ‘Cakes’.

```python
def fetch_images(driver,keyword):
    search_bar = driver.find_element(By.XPATH, '//input[@class="og3lId"]')
    search_bar.clear()
    search_bar.send_keys(keyword)
    search_bar.send_keys(Keys.RETURN)
    time.sleep(3)
    try:      
        images = driver.find_elements(By.XPATH, '//img[@class="rg_i Q4LuWd"]')
        img_urls = []
        for img in images[:10]:
            img_urls.append(img.get_attribute('src'))
            time.sleep(1)
            driver.find_element(By.XPATH, '//body').send_keys(Keys.ESCAPE)
    except NoSuchElementException :
        print("NoSuchElementException is coming")
    except StaleElementReferenceException :
        print("StaleElementReferenceException is coming")
    return img_urls
```

```python
driver = webdriver.Chrome()
driver.get("https://images.google.com/")
driver.maximize_window()

keywords = ['fruits', 'cars', 'Machine Learning', 'Guitar', 'Cakes']
search_bar = driver.find_element(By.XPATH, '//textarea[@class="gLFyf"]')
search_bar.clear()
search_bar.send_keys("google")
search_bar.send_keys(Keys.RETURN)
df = pd.DataFrame(columns=keywords)
for keyword in keywords:
    img_urls = fetch_images(driver, keyword)
    df[keyword] = img_urls
    time.sleep(3)
df
```

```python
driver.quit()
```


# 5. Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps.

```python
driver = webdriver.Chrome()
driver.get("https://www.google.com/maps")
driver.maximize_window()
```

```python
userInput = input("pls enter the location for search: ")

try:
    searchResult=driver.find_element(By.ID,"searchboxinput")
    searchResult.clear()
    searchResult.send_keys(userInput)
    searchResult.send_keys(Keys.ENTER)
except NoSuchElementException:
    print("NoSuchElementException occurs")
```

```python
def extract_coordinates_from_url(url):
    if "@" in url:
        start_index = url.find("@") + 1
        end_index = url.find(",", start_index)
        latitude = url[start_index:end_index]

        start_index = end_index + 1
        end_index = url.find(",", start_index)
        longitude = url[start_index:end_index]

        return latitude, longitude
    else:
        return None, None

# Extracting coordinates from URL
url = driver.current_url
latitude, longitude = extract_coordinates_from_url(url)
print("URL:=",url)
print(f"Coordinates for {userInput}: Latitude = {latitude}, Longitude = {longitude}")
```

```python
driver.quit()
```
