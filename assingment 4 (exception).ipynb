# 1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: 

A) Rank

B) Name

C) Artist

D) Upload date

E) Views

```python
driver = webdriver.Chrome()
driver.get(' https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
driver.maximize_window()
```

```python
# Empty list for scraping data
Rank=[]
Name =[]
Artist =[]
Upload_date =[]
Views=[]
```

```python
try:
    j=1
    rank= driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr')
    for i in rank:
        Rank.append(j)
        j= j+1
except NoSuchElementException:
    Rank.append('-')
    
    
# Extracting Name  Via Xpath
try:
    name= driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[1]')
    for i in name:
        Name.append(i.text)
except NoSuchElementException:
    Name.append('-')   
    
# Extracting Artist  Via Xpath
try:
    artist= driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[2]')
    for i in artist:
        Artist.append(i.text)
except NoSuchElementException:
    Artist.append('-')

# Extracting Upload_date  Via Xpath
try:
    upload_date= driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[3]')
    for i in upload_date:
        Upload_date.append(i.text)
except NoSuchElementException:
    Upload_date.append('-')
        
# Extracting Views  Via Xpath
try:
    views= driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[4]')
    for i in views:
        Views.append(i.text)
except NoSuchElementException :
    Views.append('-') 
```

```python
df = pd.DataFrame({'Rank':Rank,'Name':Name,'Artist':Artist,'Upload_date':Upload_date,'Views':Views})
df
```

```python
driver.quit()
```

```python
# Empty list for scraping data
Repository_title =[]
Repository_description =[]
Contributors_count =[]
Language_used=[]
```

```python
    
# Extracting Repository_title  Via Xpath
try:
    repository_title= driver.find_elements(By.XPATH,'//span[@class="text-normal"]')
    for i in repository_title:
        Repository_title.append(i.text)
except NoSuchElementException:
    Repository_title.append('-')
    
    
# Extracting Repository_description  Via Xpath
try:
    repository_description= driver.find_elements(By.XPATH,'//p[@class="col-9 color-fg-muted my-1 pr-4"]')
    for i in repository_description:
        Repository_description.append(i.text)
except NoSuchElementException:
    Repository_description.append('-')

# Extracting Contributors_count  Via Xpath
try:
    contributors_count= driver.find_elements(By.XPATH,'//div[@class="f6 color-fg-muted mt-2"]/a[2]')
    for i in contributors_count:
        Contributors_count.append(i.text)
except NoSuchElementException:
    Contributors_count.append('-')
        
# Extracting Language_used  Via Xpath
try:
    language_used= driver.find_elements(By.XPATH,'//div[@class="f6 color-fg-muted mt-2"]/span[1]')
    for i in language_used:
        if(i.text == 'Built by'):
            Language_used.append('-')
        else:
             Language_used.append(i.text)
except NoSuchElementException :
    Language_used.append('-')  
```
